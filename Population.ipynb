```python
#Dependencies
import pandas as pd
import os
```

# Importing CSVs

```python
pop_fist = os.path.join("Resources/pop-est2010.csv")
pop_fist = pd.read_csv(pop_fist)
pop_fist.head()
```

```python
pop_last = os.path.join("Resources/pop-est2020.csv")
pop_last = pd.read_csv(pop_last)
pop_last.head()
```

```python
#States Abbreviations source: https://www.ssa.gov/international/coc-docs/states.html
states_path = os.path.join("Resources/states_abbrev.csv")
states_df = pd.read_csv(states_path, encoding="ISO-8859-1")
states_df.head(60)
```

# Cleaning columns

### 2000-2010 Database

```python
# Use only totals
ag_state_2000 = pop_fist[(pop_fist['SEX'] == 0)&(pop_fist['ORIGIN']==0) & (pop_fist['RACE']==0) & (pop_fist['AGEGRP']==0)]
ag_state_2000.head()
```

```python
ag_state_2000 = ag_state_2000[['NAME', 'POPESTIMATE2000', 'POPESTIMATE2001','POPESTIMATE2002', 'POPESTIMATE2003', 'POPESTIMATE2004','POPESTIMATE2005', 'POPESTIMATE2006','POPESTIMATE2007','POPESTIMATE2008', 'POPESTIMATE2009']]
ag_state_2000.head()
```

```python
# Transpose database
ag_state_2000 = pd.melt(ag_state_2000, id_vars=["NAME"], 
            var_name="year", value_name="population")
ag_state_2000.head()
```

### 2010-2016 Database

```python
# Aggregate all age groups per state in fist database
ag_state_2020 = pop_last[['NAME','POPESTIMATE2010', 'POPESTIMATE2011','POPESTIMATE2012', 'POPESTIMATE2013', 'POPESTIMATE2014','POPESTIMATE2015', 'POPESTIMATE2016']].groupby(['NAME']).agg(['sum']
)

#Reset Index
ag_state_2020 = ag_state_2020.reset_index()
                     
ag_state_2020.head()
```

```python
# Transpose database
ag_state_2020 = pd.melt(ag_state_2020, id_vars=["NAME"], 
            var_name="year", value_name="population")
ag_state_2020.head()
```

### Join databases

```python
final_census = pd.concat([ag_state_2000, ag_state_2020], axis=0, join='outer', ignore_index=False)
final_census.head()
```

```python
# Clean Year and State column
final_census['year'] = final_census['year'].str[-4:]
final_census = final_census.rename(columns={'NAME': 'State'})
final_census['State'] = final_census['State'].str.upper()
final_census.head(60)
```

```python
# Validation count
sum_ = final_census[(final_census['year'] == '2000') &(final_census['State'] != 'UNITED STATES')]
sum_['population'].sum()
```

```python
#Change state names to abbreviation
final_census1 = pd.merge(final_census, states_df, left_on="State", right_on="state", how="left")
final_census1.head()
```

```python
final_census1 = final_census1[(final_census1['State'] != 'NORTHEAST REGION') & (final_census1['State'] != 'SOUTH REGION') & (final_census1['State'] != 'MIDWEST REGION') & (final_census1['State'] != 'WEST REGION')]
```

```python
# Validation count
sum_ = final_census1[(final_census1['year'] == '2016') &(final_census1['State'] != 'UNITED STATES')]
sum_['population'].sum()
```

```python
final_census1 = final_census1[(final_census1['State'] != 'UNITED STATES')]
final_census1.head(40)
```

```python
# Change Road Island state_code and drop extra column
final_census1 = final_census1.drop('state', 1)
final_census1.loc[final_census1.State == 'RHODE ISLAND', 'state_code'] = "RI"
final_census1.head()
```
